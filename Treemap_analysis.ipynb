```python
import pandas as pd
import numpy as np
import operator
import plotly.express as px
```

```python
FINAL_UNIV = 'Universal Music Group'
FINAL_SONY = 'Sony Music Entertainment'
FINAL_WARN = 'Warner Records'
FINAL_INDI = 'Independent'
FINAL_UNKN = 'Unknown'
FINALS = [FINAL_UNIV, FINAL_SONY, FINAL_WARN, FINAL_INDI, FINAL_UNKN]
```

```python
def create_tree_map(label_map_path, target, dataset_name='', threshold=500):
    df = pd.read_csv(label_map_path)
    df = df.loc[(df['occurrences'] >= threshold) & (df[target].isin(FINALS))]

    fig = px.treemap(df,
                     path=[px.Constant(dataset_name + " labels > 500occ"), target, 'record_label_low'],
                     values='occurrences',
                     )
    fig.update_layout(margin=dict(t=50, l=25, r=25, b=25))
    fig.show()

```

## Plot treemap for all five datasets.

```python
create_tree_map('label_map_final_mpd.csv', 'record_label_major', 'MPD')
```

```python
create_tree_map('label_map_final_lfm.csv', 'record_label_major', 'LFM')
```
